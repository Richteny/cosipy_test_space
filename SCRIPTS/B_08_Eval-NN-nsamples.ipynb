```python
sample_size = [100, 200, 300, 400, 500, 1000, 1500, 2000]
exps = 10

for size in sample_size:
    for i in range(exps):
        with open(path+f"model_experiment_{size}v{i}_samples.pkl", "rb") as f:
            exp = pickle.load(f)
            for var in ["MB","TSLA","ALB"]:
                for metric in ["MAE","RMSE","R2"]:
                    dic_results[size][var][metric].append(exp["metrics"][var][metric])
```

```python
sample_sizes = sorted(dic_results.keys())
components = ['MB', 'TSLA', 'ALB']
metrics = ['MAE', 'RMSE', 'R2']
colors = {'MAE': 'blue', 'RMSE': 'orange', 'R2': 'green'}

def extract_metric(component, metric):
    return [dic_results[size][component][metric] for size in sample_sizes]

plt.rcParams.update({'font.size': 22})
fig, axs = plt.subplots(3, 1, figsize=(12, 18), dpi=300, sharex=True)
fig.subplots_adjust(hspace=0.5) 
offset = 25  # spacing between MAE, RMSE, R²
width = 25

for i, component in enumerate(components):
    ax = axs[i]
    ax2 = ax.twinx()
    ax2.spines['right'].set_color('darkgreen')
    ax2.yaxis.label.set_color('darkgreen')
    ax2.tick_params(axis='y', colors='darkgreen')

    mae_data = [dic_results[size][component]['MAE'] for size in sample_sizes]
    rmse_data = [dic_results[size][component]['RMSE'] for size in sample_sizes]
    r2_data = [dic_results[size][component]['R2'] for size in sample_sizes]

    pos_mae = [x - offset for x in sample_sizes]
    pos_rmse = sample_sizes
    pos_r2 = [x + offset for x in sample_sizes]

    # Medians for connecting lines
    mae_meds = [np.median(vals) for vals in mae_data]
    rmse_meds = [np.median(vals) for vals in rmse_data]
    r2_meds = [np.median(vals) for vals in r2_data]

    # Plot lines behind the boxplots
    ax.plot(pos_mae, mae_meds, color=colors['MAE'], linestyle='-', linewidth=1, zorder=1)
    ax.plot(pos_rmse, rmse_meds, color=colors['RMSE'], linestyle='-', linewidth=1, zorder=1)
    ax2.plot(pos_r2, r2_meds, color=colors['R2'], linestyle='--', linewidth=1, zorder=1)

    # Plot MAE & RMSE boxplots
    ax.boxplot(mae_data, positions=pos_mae, widths=width, patch_artist=True,
               boxprops=dict(facecolor=colors['MAE'], alpha=0.6),
               medianprops=dict(color='black'),
               flierprops=dict(marker='o', alpha=0.3, markersize=2))
    
    ax.boxplot(rmse_data, positions=pos_rmse, widths=width, patch_artist=True,
               boxprops=dict(facecolor=colors['RMSE'], alpha=0.6),
               medianprops=dict(color='black'),
               flierprops=dict(marker='x', alpha=0.3, markersize=2))

    # Plot R² boxplots
    ax2.boxplot(r2_data, positions=pos_r2, widths=width, patch_artist=True,
                boxprops=dict(facecolor=colors['R2'], alpha=0.6),
                medianprops=dict(color='black'),
                flierprops=dict(marker='s', alpha=0.3, markersize=2))

    # Axis labels and title
    #ax.set_title(f'{component}')
    #ax.set_xlabel('Training Sample Size')
    if component in ["TSLA", "ALB"]:
        ax.set_ylabel("")
    else:
        ax.set_ylabel('MAE / RMSE')
    if component in ["MB","TSLA"]:
        ax2.set_ylabel("")
    else:
        ax.set_xlabel('Training Sample Size')
    ax2.set_ylabel('R² Score', color="darkgreen")
    ax2.set_yticks(np.arange(0.92,1.0+0.02,0.02))
    ax2.set_ylim(0.92, 1.0)
    ax.set_ylabel('MAE / RMSE')
    ax.grid(True)

# Set x-ticks only on the primary axis
for ax in axs:
    ax.set_xticks(sample_sizes)
    ax.set_xticklabels(sample_sizes, rotation=45)
#else:
#    ax.set_xticklabels([])

#
fig.text(0.01, 0.992, 'a) Mass Balance', transform=fig.transFigure, fontsize=24, va='top')
fig.text(0.01, 0.692, 'b) SLA', transform=fig.transFigure, fontsize=24, va='top')
fig.text(0.01, 0.38, r'c) $\bar{\alpha}$', transform=fig.transFigure, fontsize=24, va='top')

# Manual legend
legend_elements = [
    Line2D([0], [0], color=colors['MAE'], lw=4, label='MAE'),
    Line2D([0], [0], color=colors['RMSE'], lw=4, label='RMSE'),
    Line2D([0], [0], color=colors['R2'], lw=4, linestyle='--', label='R²'),
]
axs[2].legend(handles=legend_elements, loc='center right')
fig.tight_layout()
#plt.suptitle('NN Performance by Training Sample Size', fontsize=16)
#plt.tight_layout(rect=[0, 0, 1, 0.95])
#plt.show()
plt.savefig("/mnt/C4AEBBABAEBB9500/OneDrive/PhD/PhD/Data/Hintereisferner/Figures/emulator_samplesize_test.pdf", bbox_inches="tight")
```
